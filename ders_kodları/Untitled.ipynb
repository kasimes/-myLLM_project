```python
import numpy as np
import matplotlib.pyplot as plt

# Veri
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label="sin(x)")
plt.title("Matplotlib Örneği: Sinüs Eğrisi")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.legend()


plt.show()
```
